## Reading Image 

```python
img = cv2.imread('hawk.png')
```

```python
img = cv2.normalize(img.astype('float'), None, 0.0, 1.0, cv2.NORM_MINMAX)
```

```python
img = cv2.resize(img, (130, 120))
```

```python
print(img.shape)
plt.imshow(img)
```

## Arithmetic Mean Filter 

```python
# Arithmetic Mean Filter
filtered_img = np.zeros(img.shape)
n = 3
m = 3
filter = np.ones((n, m))/(n*m)
sum = 0
for i in range(3):
    for j in range((n//2), img.shape[0]-(n//2)):
        for k in range((m//2), img.shape[1]-(m//2)):
                    filtered_img[j, k, i] = np.sum(img[j-(n//2):j+(n//2)+1, k-(m//2):k+(m//2)+1, i] * filter)
```

```python
plt.imshow(filtered_img)
```

## Geometric Mean Filter

```python
#Geometric Mean Filter
g_filtered = np.ones(img.shape)
n = 3
m = 3
for i in range(3):
    for j in range((n//2), img.shape[0]-(n//2)):
        for k in range((m//2), img.shape[1]-(m//2)):
                    g_filtered[j, k, i] = np.power(np.prod(img[j-(n//2):j+(n//2)+1, k-(m//2):k+(m//2)+1, i]), 1/(n*m))
```

```python
plt.imshow(g_filtered)
```
